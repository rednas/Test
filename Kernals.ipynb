```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

```
#kernel = np.ones((3,3),np.float32)/25
verticalkernel = np.float32([[-1,2,-1],[-1,2,-1],[-1,2,-1]])

dst = cv2.filter2D(image,-1,verticalkernel)
cv2_imshow(dst)
```

```
#kernel = np.ones((3,3),np.float32)/25
horizontalkernel = np.float32([[-1,-1,-1],[2,2,2],[-1,-1,-1]])

dst = cv2.filter2D(image,-1,horizontalkernel)
cv2_imshow(dst)
```

```

inclinedkernel = np.float32([[-1,-1,2],[-1,2,-1],[2,-1,-1]])

dst = cv2.filter2D(image,-1,inclinedkernel)
cv2_imshow(dst)
```

```
sharpeningkernel = np.float32([[0,-0.25,0],[-0.25,2,-0.25],[0,-0.25,0]])

dst = cv2.filter2D(image,-1,sharpeningkernel)
cv2_imshow(dst)
```

```
blurringkernel = np.float32([[0,1/16,0],[1/16,4,1/16],[0,1/16,0]])

dst = cv2.filter2D(image,-1,blurringkernel)
cv2_imshow(dst)
```

```
indentitykernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])

dst = cv2.filter2D(image,-1,indentitykernel)
cv2_imshow(dst)
```
